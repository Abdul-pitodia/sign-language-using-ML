```python
directory = 'D:/Abdul 2020/Study/Coding Practice/be proj code/new_data/z'

cam = cv2.VideoCapture(0)

cv2.namedWindow("test")

while True:
    ret, frame = cam.read()
    if not ret:
        print("failed to grab frame")
        break
    
    blur = cv2.GaussianBlur(frame,(5,5),0)
    blur = cv2.medianBlur(blur, 11)
    blur = blur[0:370,0:370]

    imgray = cv2.cvtColor(blur,cv2.COLOR_BGR2GRAY)
    imgraycopy = imgray.copy()

    th3 = cv2.adaptiveThreshold(imgraycopy, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY_INV,11,2)
    _,mask = cv2.threshold(th3, 80, 255, cv2.THRESH_BINARY_INV+cv2.THRESH_OTSU)
        
    cv2.imshow("test", mask)

    k = cv2.waitKey(1)
    if k%256 == 27:
        # ESC pressed
        print("Escape hit, closing...")
        break
        
    elif k%256 == 32:
        # SPACE pressed
        for i in range(1, 1001):
            img_name = "{}.png".format(i)
#             time.sleep(0.1)
            cv2.imwrite(os.path.join(directory, img_name), mask)
            print("{} written!".format(img_name))

cam.release()
cv2.destroyAllWindows()
```

```python
DATADIR = "D:/Abdul 2020/Study/Coding Practice/be proj code/new_data"
lst = list(string.ascii_lowercase)

#Array of training image arrays
training_data = []

def create_training_data():
    for ind in lst:
        path = os.path.join(DATADIR, ind)
        print(path)
        for img in os.listdir(path):
            img_array = cv2.imread(os.path.join(path,img),cv2.IMREAD_GRAYSCALE)
            img_array = cv2.resize(img_array,(100,100))
            training_data.append([img_array, ind])
#             break
#         break

create_training_data()
```
